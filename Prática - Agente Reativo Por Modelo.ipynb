Os agentes reativos baseados em modelos são agentes que podem tomar uma decisão de forma mais acurada. Neste trabalho você deverá implementar um novo agente para jogar o jogo da velha contra o agente reativo simples implementado anteriormente.


1.   O jogo será máquina (agente reativo baseado em modelo) vs máquina (agente reativo simples)
1.  Você deverá implementar uma estratégia inteligente de escolha da jogada que utilize o estado passado para determinar aonde jogar.
2.  A primeira jogada será realizada pelo agente reativo simples
1.  Não é obrigatório implementar interface gráfica.
2.  Ao final do jogo deverá salvar quem ganhou e quem perdeu.
1.   Você deverá executar o jogo 50 vezes e plotar um gráfico com o número de vitórias de cada um dos agentes e o número de empates


```python
import random
import matplotlib.pyplot as plt

tab = [0, 1, 2, 3, 4, 5, 6, 7, 8]

def desenho():
    print(tab[0], '|', tab[1], '|', tab[2])
    print('-' * 9)
    print(tab[3], '|', tab[4], '|', tab[5])
    print('-' * 9)
    print(tab[6], '|', tab[7], '|', tab[8])


def modelo():
    quatro = False
    
    print('Jogada agente baseado em modelo:')
    opcoesValidas = []
    while(tab[4] == 4):
        opcoesValidas.append(4)
        tab[4] = 'O'
        quatro = True
        break
        
    for i in range(0, 9):
        if quatro != True:
            if tab[0] != 'X' and tab[0] != 'O' or tab[2] != 'X' and tab[2] != 'O' or tab[6] != 'X' and tab[6] != 'O' or tab[8] != 'X' and tab[8] != 'O':
                opc = 0,2,6,8
                opcoesValidas.extend(opc)
                opcao = random.choice(opcoesValidas)
                tab[opcao] = 'O' 
        else: 
            if quatro != True:
                if tab[1] != 'X' and tab[1] != 'O' or tab[3] != 'X' and tab[3] != 'O' or tab[5] != 'X' and tab[5] != 'O' or tab[7] != 'X' and tab[7] != 'O':
                    opcoesValidas = [1,3,5,7]
                    opcao = random.choice(opcoesValidas)
                    tab[opcao] = 'O' 
        break
        
        
def jogadaIA():
    print('Jogada agente simples:')
    opcoesValidas = []
    for i in range(0, 9):
        if tab[i] != 'X' and tab[i] != 'O':
            opcoesValidas.append(i)
    opcao = random.choice(opcoesValidas)
    tab[opcao] = 'X'

def ganhou(turno):
      if(
          (tab[0] == turno and tab[1] == turno and tab[2] == turno) or
          (tab[3] == turno and tab[4] == turno and tab[5] == turno) or
          (tab[6] == turno and tab[7] == turno and tab[8] == turno) or
          (tab[0] == turno and tab[3] == turno and tab[6] == turno) or
          (tab[1] == turno and tab[4] == turno and tab[7] == turno) or
          (tab[2] == turno and tab[5] == turno and tab[8] == turno) or
          (tab[0] == turno and tab[4] == turno and tab[8] == turno) or
          (tab[2] == turno and tab[4] == turno and tab[6] == turno)
      ):
        return True
      else:
        return False

desenho()

turno = True
contar_modelo = 0
contar_simples = 0
contar_velha = 0
contar = 0

while(contar <50):
    for i in range(0, 9):
        if turno:
            jogadaIA()
            if ganhou('X'):
                desenho()
                print('Agente simples ganhou!')
                contar_simples += 1 
                break
        else:            
            modelo()
            if ganhou('O'):
                desenho()
                print('Agente baseado em modelo ganhou!')
                contar_modelo += 1 
                break
        desenho()
        turno = not turno
        if i == 8:
            print('Deu velha')
            print('Fim do jogo.')
            contar_velha +=1
            break
    contar = contar + 1
    
print('Agente baseado em modelo ganhou: ', contar_modelo)
print('Agente simples ganhou: ', contar_simples)
print('Velha: ', contar_velha)

titulo = ['modelo','simples','velha']
dados = [contar_modelo, contar_simples, contar_velha]
plt.ylabel('Qtd vitorias')
plt.bar(titulo, dados, color='green')
```
